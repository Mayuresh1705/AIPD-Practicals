```python
import pandas as pd

df1 = pd.DataFrame({"Experience": [2,5,3,7,1,6,4,8],"Gender": [0,0,0,0,1,1,1,1],"Hired": [1,1,1,1,0,0,0,0]})
```

```python
df1
```

```python
x=df1[["Experience","Gender"]]
y=df1["Hired"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
pred1=lr.predict(x_test)
```

```python
pred1
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(pred1,y_test)
acc
```

```python
new_data = pd.DataFrame({
    "Experience": [5],
    "Gender": [0]
})

prediction = lr.predict(new_data)

print("Prediction:", prediction[0])
```

```python
df=pd.read_csv("/content/drive/MyDrive/College/SEM-6/AIPD/Dataset/recruitment_data.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
X = df.drop("HiringDecision", axis=1)
Y = df["HiringDecision"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score,ConfusionMatrixDisplay,confusion_matrix,classification_report
acc_lr=accuracy_score(y_pred,y_test)
```

```python
acc_lr
```

```python
conf_mat = confusion_matrix(y_test, y_pred)
cm_display = ConfusionMatrixDisplay(confusion_matrix=conf_mat, display_labels=lr.classes_)
cm_display.plot()
```

```python
cr=classification_report(y_pred,y_test)
print(cr)
```
