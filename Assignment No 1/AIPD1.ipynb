```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("../admin1/Downloads/diabetes.csv")
```

```python
x=df.drop(columns=['Outcome'])
y=df['Outcome']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred1=lr.predict(x_test)
```

```python
y_pred1
```

```python
from sklearn.metrics import accuracy_score
acc1=accuracy_score(y_pred1,y_test)
acc1
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_pred1,y_test)
print(cr)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(y_pred1,y_test)
displaycm=ConfusionMatrixDisplay(cm)
displaycm.plot()
```

#### Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
gb=GaussianNB()
```

```python
gb.fit(x_train,y_train)
```

```python
y_pred2=gb.predict(x_test)
y_pred2
```

```python
from sklearn.metrics import accuracy_score
acc2=accuracy_score(y_pred2,y_test)
acc2
```

```python
from sklearn.metrics import classification_report
cr=classification_report(y_pred2,y_test)
print(cr)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cm=confusion_matrix(y_pred2,y_test)
displaycm=ConfusionMatrixDisplay(cm)
displaycm.plot()
```

```python
import matplotlib.pyplot as plt

models = ['Logistic Regression', 'Gaussian NB']
accuracies = [acc1,acc2]

plt.figure(figsize=(10,8))
plt.bar(models, accuracies, color=['red', 'blue'])
plt.xlabel("Models")
plt.ylabel("Accuracy")
plt.ylim(0, 1) 
plt.show()
```
